# Visualisation Of CIRFAR10 FAR/NEAR OOD Dataset

```python
import warnings
warnings.filterwarnings("ignore")

import torchvision
import matplotlib.pyplot as plt
from torchvision import transforms

from dataset import CIFAR10_NEAR_ODD_Dataset, CIFAR10_FAR_ODD_Dataset, CustomDataset
```

## Visualisation Of CIRFAR10 NEAR OOD Dataset

- Labels 0-9 are in-distribution data from CIFAR10
- Labels 10-109 are near out-of-distribution data from CIFAR100

```python
transform_train = transforms.Compose([
    torchvision.transforms.Resize(64),
    torchvision.transforms.RandomCrop(64),
    torchvision.transforms.ToTensor()
])

transform_test = transforms.Compose([
    torchvision.transforms.Resize(64),
    torchvision.transforms.ToTensor()
])

train_dataset = CIFAR10_NEAR_ODD_Dataset(is_train=True, transform=transform_train)
test_dataset = CIFAR10_NEAR_ODD_Dataset(is_train=False, transform=transform_test)
print("Training Data Size:", len(train_dataset))
print("Testing Data Size:", len(test_dataset))
print("Input Image Shape:", train_dataset[0][0].shape)

cifar10_dataset = CustomDataset(dataset_name="cifar10", is_train=True)
cifar100_dataset = CustomDataset(dataset_name="cifar100", is_train=True)
cifar10_near_ood_dataset = CIFAR10_NEAR_ODD_Dataset(is_train=True)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
img, label = cifar10_dataset[0]
axes[0].imshow(img)
axes[0].set_title(f"Label {label}")
img, label = cifar10_near_ood_dataset[0]
axes[1].imshow(img)
axes[1].set_title(f"Label {label}")
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
img, label = cifar100_dataset[0]
axes[0].imshow(img)
axes[0].set_title(f"Label {label}")
img, label = cifar10_near_ood_dataset[len(cifar10_dataset)]
axes[1].imshow(img)
axes[1].set_title(f"Label {label}")
plt.show()
```

## Visualisation Of CIRFAR10 FAR OOD Dataset

- Labels 0-9 are in-distribution data from CIFAR10
- Labels 10-19 are far out-of-distribution data from MNIST
- Labels 20-29 are far out-of-distribution data from FashionMNIST
- Labels 30-39 are far out-of-distribution data from NOTMNIST

```python
transform_train = transforms.Compose([
    torchvision.transforms.Resize(64),
    torchvision.transforms.RandomCrop(64),
    torchvision.transforms.ToTensor()
])

transform_test = transforms.Compose([
    torchvision.transforms.Resize(64),
    torchvision.transforms.ToTensor()
])

train_dataset = CIFAR10_FAR_ODD_Dataset(is_train=True, transform=transform_train)
test_dataset = CIFAR10_FAR_ODD_Dataset(is_train=False, transform=transform_test)
print("Training Data Size:", len(train_dataset))
print("Testing Data Size:", len(test_dataset))
print("Input Image Shape:", train_dataset[0][0].shape)

cifar10_dataset = CustomDataset(dataset_name="cifar10", is_train=True)
mnist_dataset = CustomDataset(dataset_name="mnist", is_train=True)
fashionmnist_dataset = CustomDataset(dataset_name="fashionmnist", is_train=True)
notmnist_dataset = CustomDataset(dataset_name="notmnist", is_train=True)
cifar10_far_ood_dataset = CIFAR10_FAR_ODD_Dataset(is_train=True)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
img, label = cifar10_dataset[0]
axes[0].imshow(img)
axes[0].set_title(f"Label {label}")
img, label = cifar10_far_ood_dataset[0]
axes[1].imshow(img)
axes[1].set_title(f"Label {label}")
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
img, label = mnist_dataset[0]
axes[0].imshow(img)
axes[0].set_title(f"Label {label}")
img, label = cifar10_far_ood_dataset[len(cifar10_dataset)]
axes[1].imshow(img)
axes[1].set_title(f"Label {label}")
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
img, label = fashionmnist_dataset[0]
axes[0].imshow(img)
axes[0].set_title(f"Label {label}")
img, label = cifar10_far_ood_dataset[len(cifar10_dataset) + len(mnist_dataset)]
axes[1].imshow(img)
axes[1].set_title(f"Label {label}")
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
img, label = notmnist_dataset[0]
axes[0].imshow(img)
axes[0].set_title(f"Label {label}")
img, label = cifar10_far_ood_dataset[len(cifar10_dataset) + len(mnist_dataset) + len(fashionmnist_dataset)]
axes[1].imshow(img)
axes[1].set_title(f"Label {label}")
plt.show()
```
